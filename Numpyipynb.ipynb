```python


# Create a 3D array
# This array has 2 'layers' (matrices)
# Each layer has 2 rows
# Each row has 3 elements
d = np.array([
    [ # First layer
        [1, 2, 3],
        [4, 5, 6]
    ],
    [ # Second layer
        [7, 8, 9],
        [10, 11, 12]
    ]
])

print(d)
print(f"Shape of array d: {d.shape}") # Should output (2, 2, 3)
print(f"Number of dimensions: {d.ndim}")
```

Reorganizing Arrays

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)
```

```python
#If we wanna make it 8*1
after = before.reshape((8,1))
after
```

```python
after = before.reshape((4,2))
after
```

```python
after = before.reshape((2,2,2))
after
```

```python
# Vertically stacking vectors
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])
```

```python
np.vstack([v1,v2])
```

```python
#Horizontal stack
h1 = np.ones((2,4))
h2 = np.ones((2,2))
```

```python
np.hstack((h1,h2))
```
